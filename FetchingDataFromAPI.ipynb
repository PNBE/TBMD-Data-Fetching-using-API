```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]      # we require only specific column of the given data frame
temp_df
```

Now we have to perform this in loop. The number of times the loop will run is equal to the number of pages. In each loop we made a new data frame and then append this data frame with the older one.

```python
df = pd.DataFrame()     # empty data frame
df
```

```python
for i in range(1,429):           # it contain 428 pages
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=6730a21937e1868214e2da4e0df38041&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)             # using ignore_index = True, we can keep the data in sequiential form while appending
```
